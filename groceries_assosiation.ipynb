```python
# pivot table is a way to reshape and summarize the data
# create baskets
baskets = data.pivot_table(index="Member_number", columns="itemDescription", aggfunc='size', fill_value=0)
baskets
```

```python
# CReate an encode function to encode the pivot table
# This is basically yes or no answer to whether a production was purchased or not in a transaction
# in other words, it is whther a product exists in an itemset of a transaction or not
# if yes, we replace with 1 and other we have 0

def encode(x):
    # x is the quantity as shown on the pivot
    if x<=0:
        return 0
    if x > 0:
        return 1
    
# we shall then apply the above encoding function to the baskets
baskets_sets = baskets.applymap(encode)
baskets_sets
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
frequent_items = apriori(baskets_sets, min_support=0.05, use_colnames=True)
rules = association_rules(frequent_items, metric='lift')
rules
```

```python
rules[(rules['lift'] > 1) & (rules['confidence'] > 0.4)]
```
