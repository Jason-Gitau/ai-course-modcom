```python
#load/read data from csv file
import pandas as pd
# update pandas  to show all columns
pd.set_option('display.max_columns', None)


data = pd.read_csv('datasets/bank.csv')



data
```

```python
# Identify numeric columns (adjust these based on your actual column names)
numeric_cols = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']

# Convert to numeric, coercing errors to NaN
data[numeric_cols] = data[numeric_cols].apply(pd.to_numeric, errors='coerce')

# Show data types after conversion
data.dtypes
```

```python
# Fill numeric columns with median
for col in numeric_cols:
    data[col].fillna(data[col].median(), inplace=True)

# Identify categorical columns
categorical_cols = ['Gender', 'Married', 'Education', 'Self_Employed', 'Property_Area', 'Loan_Status']

# Fill categorical columns with mode (most frequent value)
for col in categorical_cols:
    if data[col].dtype == 'object':  # Only fill string columns
        data[col].fillna(data[col].mode()[0], inplace=True)

# Verify no missing values remain
data.isnull().sum()
```

```python
# Standardize categorical values (title case and consistent values)
data['Gender'] = data['Gender'].str.title().replace({'Male': 'Male', 'Female': 'Female'})
data['Married'] = data['Married'].str.title().replace({'Yes': 'Yes', 'No': 'No'})
data['Education'] = data['Education'].str.title().replace({'Graduate': 'Graduate', 'Not Graduate': 'Not Graduate'})
data['Self_Employed'] = data['Self_Employed'].str.title().replace({'Yes': 'Yes', 'No': 'No'})
data['Loan_Status'] = data['Loan_Status'].str.title().replace({'Y': 'Yes', 'N': 'No'})

# Show sample of cleaned categorical data
data[categorical_cols].head()
```

```python
# do we have any empty record?
data.isnull().sum()#shows the missing records per columns
# if the missing is less than 2% then that should be ok
```

```python
# incase we choose to fill the empty we use a relative value eg 
# for numeric column we use the average so this way the data does not deviate

data.describe()#only works with  all numeric values
```

```python
# Ensure Loan_ID has no missing values (primary key)
data.dropna(subset=['Loan_ID'], inplace=True)

# Reset index after cleaning
data.reset_index(drop=True, inplace=True)

# Display final cleaned data info
print("Final Data Info:")
print(data.info())

print("\nFirst 5 rows:")
data.head()
```
